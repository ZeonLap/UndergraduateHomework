```python
import matplotlib.pyplot as plt
import csv
```

```python
train_loss = []

with open('CNN_trainloss.csv') as file:
    csv_reader = csv.reader(file, delimiter=',')
    columns = zip(*csv_reader)
    for index, column in enumerate(columns):
        if column[0] == 'cnn_dr_0.5 - train loss':
            train_loss.append(column[1:])

train_loss = [float(x) for x in train_loss[0]]

val_loss = []

with open('CNN_valloss.csv') as file:
    csv_reader = csv.reader(file, delimiter=',')
    columns = zip(*csv_reader)
    for index, column in enumerate(columns):
        if column[0] == 'cnn_dr_0.5 - val loss':
            val_loss.append(column[1:])

val_loss = [float(x) for x in val_loss[0]]
```

```python
plt.plot(range(1, 101), train_loss, label='train loss')
plt.plot(range(1, 101), val_loss, label='val loss')
plt.title('CNN')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend()
plt.show()
```
