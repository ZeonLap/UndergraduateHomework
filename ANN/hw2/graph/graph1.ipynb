```python
import matplotlib.pyplot as plt
import csv
```

```python
data = []

with open('CNN_testloss.csv') as file:
    csv_reader = csv.reader(file, delimiter=',')
    columns = zip(*csv_reader)
    for index, column in enumerate(columns):
        if (index > 0) and (index % 3 == 1):
            data.append(float(column[-1]))

data.reverse()
data
```

```python
labels = [i / 10 for i in range(1, 10)]
plt.bar([1, 2, 3, 4, 5, 6, 7, 8, 9], data, tick_label=labels)
plt.title('CNN')
plt.xlabel('dropout rate')
plt.ylabel('test_loss')
# plt.ylim(0.4, 1.6)
plt.show()
```
